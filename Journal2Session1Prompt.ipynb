```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)

def get_exp(val):
    return np.exp(val)
```

```python
y = get_exp(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y)

plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')
plt.title('Exponential Growth of Awesomeness')

plt.savefig('awesomeness_plot.pdf')

plt.show()
```
